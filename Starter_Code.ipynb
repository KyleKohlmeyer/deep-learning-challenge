## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
# application_df = application_df.drop(columns=["EIN", "NAME"])
# application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN'.
application_df = application_df.drop(columns=["EIN"])
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_cat = application_df.dtypes[application_df.dtypes == "object"].index.tolist()
application_df[application_cat].nunique()
```

```python
# Look at NAME value counts for binning
application_df["NAME"].value_counts()
```

```python
# Retrieve a list of all company names with value_count below chosen cutoff value of 5
below_cutoff_name = application_df["NAME"].value_counts().loc[lambda x : x<5].index.tolist()
below_cutoff_name
```

```python
# use the variable name `application_names_to_replace`
application_names_to_replace = below_cutoff_name

# Replace in dataframe
for app in application_names_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(app,"Other")

# Check to make sure binning was successful
application_df['NAME'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df["APPLICATION_TYPE"].value_counts()
```

```python
# Retrieve a list of all classifications with value_count below chosen cutoff value of 500
below_cutoff_app_type = application_df["APPLICATION_TYPE"].value_counts().loc[lambda x : x<500].index.tolist()
below_cutoff_app_type
```

```python
# use the variable name `application_types_to_replace`
application_types_to_replace = below_cutoff_app_type

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df["CLASSIFICATION"].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
application_df["CLASSIFICATION"].value_counts().loc[lambda x : x>1]
```

```python
# Retrieve a list of all classifications with value_count below chosen cutoff value of 1000
below_cutoff_class = application_df["CLASSIFICATION"].value_counts().loc[lambda x : x<1000].index.tolist()
below_cutoff_class
```

```python
# use the variable name `classifications_to_replace`
classifications_to_replace = below_cutoff_class

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies WITHOUT 'NAME'`
# cat_dummies = pd.get_dummies(application_df, columns=['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS'], drop_first=True)
# cat_dummies.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies WITH 'NAME'`
cat_dummies = pd.get_dummies(application_df, columns=['NAME','APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS'], drop_first=True)
cat_dummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = cat_dummies.IS_SUCCESSFUL.values
X = cat_dummies.drop(columns="IS_SUCCESSFUL").values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Get number of columns
num_columns = X.shape[1]
print("Number of columns:", num_columns)
```

```python
import keras
keras.backend.clear_session()
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=100, activation="relu", input_dim=num_columns))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=30, activation="sigmoid"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=10, activation="sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=500)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export the model to HDF5 file
nn.save('AlphabetSoupCharity_Optimized.h5')
```
